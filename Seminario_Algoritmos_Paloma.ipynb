Pseudocódigo:

Paso1: calculo todas las permutaciones de los 9 números cogidos de 5 en 5

Paso2: calculo todas las permutaciones de los 4 signos de operación

Paso3: hago una array con todas las combinaciones de cada miembro de permutaciones de números con cada uno de los miembros de las permutaciones de signos de operación

Paso4: les aplico la función EVAL(expresión)

Paso5: si nos da un número entero (ya sea positivo o negativo) lo guardamos en un array de soluciones tipo   4 = 4+2-6/3*1

Paso6: ordeno los números del array de soluciones de menor a mayor


```
# Paso1: calculo todas las permutaciones de los 9 números cogidos de 5 en 5
# Paso2: calculo todas las permutaciones de los 4 signos de operación

import numpy as np
from itertools import permutations, combinations

numeros = [1,2,3,4,5,6,7,8,9]
operadores = ['+','-','*','/']

num_permutations = [p for p in permutations(numeros,5)]
op_permutations = [p for p in permutations(operadores)]

print('\nLas combinaciones de los 9 números cogidos de 5 en 5 son: ')
print(num_permutations)
print('\nLas combinaciones de operadores son: ')
print(op_permutations)

```

```
# Paso3: hago una array con todas las combinaciones de cada miembro de permutaciones de números con cada uno de los miembros de las permutaciones de signos de operación
# Paso4: les aplico la función EVAL(expresión)

# Ahora vamos a combinar los números con los signos
i = 1
j = 1
combinaciones = []
resultados = {}
for i in num_permutations:
  for j in op_permutations:
    expresion=[]
    for nr in range(len(j)):
        expresion.append(i[nr])
        expresion.append(j[nr])
    expresion.append(i[-1])
    expres = ''.join(map(str,expresion))

    # evaluo la expresión
    res = eval(expres) 

    # me quedo solo con los resultados en los que la expresión es un valor entero, tanto positivo como negativo,
    if res % 1 == 0:
      if res in resultados.keys():
        resultados[res].append(expres) 
      else:   
        resultados[res] = [expres]

#  Ahora calculo el mínimo y el máximo

print('\nEl entero mínimo que podemos encontrar es: ', min(resultados.keys()))
print('\nEl entero máximo que podemos encontrar es: ', max(resultados.keys()))
```

```
# Vamos a listar todos los resultados de los enteros encontrados, 
# Podemos comprobar que entre ese mínimo y el máximo sí es posibe encontrar todos los enteros

def get_solution(resultados, n):
  if n in resultados.keys():
    return resultados[n]
  else:
    return None

for i in sorted(resultados.keys()):
  print(i)  
```
